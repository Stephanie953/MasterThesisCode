### Functions

```python
### Function to flatten nested lists ###

flatten = lambda l: [item for sublist in l for item in sublist]

### Function to change the items to numbers ###
def changeItem(item):
    values = {"Unimportant": 0, "Moderately Important": 1, "Important": 2, "Very Important": 3, "Paramount": 4}
    
    if item not in values.keys():
        return np.nan
    
    return values[item]

### Function to reverse the items which were stated as a negation ###

def reverse(item):
    if item == 0:
        return 4
    
    elif item == 1:
        return 3
    
    elif item == 2:
        return 2
    
    elif item == 3:
        return 1
    
    elif item == 4:
        return 0
    
    else:
        return np.nan
    
### Function to calculate Cronbach Alpha ###

def CronbachAlpha(itemscores):
    itemscores = np.asarray(itemscores)
    itemvars = itemscores.var(axis=0, ddof=1)
    tscores = itemscores.sum(axis=1)
    nitems = itemscores.shape[1]

    return (nitems / (nitems-1)) * (1 - (itemvars.sum() / tscores.var(ddof=1)))
    
```

### Cleaning & Transformation

```python
### Reset column names to the to be measured variables ###

oldVariables = data.keys()[12:54]

principles_goals = ["Reuse of artifacts", "Automation", "Integrated end-to-end thinking", "Short   cycles   and   incremental change",
            "Analytics as code", "Testing", "Monitoring", "Data-driven improvement", "Process-oriented data pipelines",
            "Continuous improvement", "Orchestration", "Empowerment of citizen users","Agility and speed","Collaboration and trust"]

newVariables = [["{} 1".format(item), "{} 2".format(item), "{} 3".format(item)] for item in principles_goals]

newVariables = flatten(newVariables)

data = data.rename(columns = {oldVariables[i]: newVariables[i] for i in range(len(oldVariables))})


### Transform data to a numerical format to analyse ### 

for key in data.keys()[12:54]:
    data[key] = data[key].apply(lambda x: changeItem(x))
    
    
list_of_reversed_columns = ['Reuse of artifacts 3','Automation 3','Integrated end-to-end thinking 3',
                           'Testing 3','Monitoring 3','Data-driven improvement 3','Process-oriented data pipelines 3',
                           'Continuous improvement 3','Orchestration 3','Empowerment of citizen users 3',
                           'Agility and speed 3','Collaboration and trust 2']

for column in list_of_reversed_columns:
    data[column] = data[column].apply(lambda x: reverse(x))
    
    
### Transform timestamp to date type ###

data["Timestamp"] = pd.to_datetime(data["Timestamp"])

#Drop the empty responses
data = data.dropna(axis = 0, how="any", thresh = 8)
```

```python
### Fill up the empty values for the principles to enhance the dataset ###

principles = ['Reuse of artifacts 1', 'Reuse of artifacts 2', 'Reuse of artifacts 3',
       'Automation 1', 'Automation 2', 'Automation 3',
       'Integrated end-to-end thinking 1', 'Integrated end-to-end thinking 2',
       'Integrated end-to-end thinking 3',
       'Short   cycles   and   incremental change 1',
       'Short   cycles   and   incremental change 2',
       'Short   cycles   and   incremental change 3', 'Analytics as code 1',
       'Analytics as code 2', 'Analytics as code 3', 'Testing 1', 'Testing 2',
       'Testing 3', 'Monitoring 1', 'Monitoring 2', 'Monitoring 3',
       'Data-driven improvement 1', 'Data-driven improvement 2',
       'Data-driven improvement 3', 'Process-oriented data pipelines 1',
       'Process-oriented data pipelines 2',
       'Process-oriented data pipelines 3']

data[principles] = data[principles].fillna(data[principles].mean())
```

### Testing internal consistency based upon Cronbach Alpha

```python
### Calculate Cronbach Alpha ###

principles_goals = ["Reuse of artifacts", "Automation", "Integrated end-to-end thinking", "Short   cycles   and   incremental change",
            "Analytics as code", "Testing", "Monitoring", "Data-driven improvement", "Process-oriented data pipelines",
            "Continuous improvement", "Orchestration", "Empowerment of citizen users","Agility and speed","Collaboration and trust"]

cronbachalphas = []

#Gather the items used to calculate the initial cronbach alpha
for item in principles_goals:    
    small_data = data[[key for key in data.keys() if item in key]].dropna(axis = 0, how = "any")
    alpha = CronbachAlpha(small_data[[key for key in data.keys() if item in key]])
    cronbachalphas.append(alpha)
    
print(pd.DataFrame.from_dict({"Items": principles_goals, "Cronbach alpha": cronbachalphas}))    
```

```python
### Correlation Heatmaps ###

# Here we create a figure instance, and 14 subplots in order to illustrate which items correlate to one another.
# This is used to improve the Cronbach Alpha by erasing items that are (too) little correlated.

# Set up complete figure
fig = plt.figure()
sns.set(context = "paper", style = "whitegrid", font_scale = 1.5)

# Initialize plots
axes = [fig.add_subplot(5,3,i) for i in range(1,15)]

# Create heatmap per item

for i in range(len(principles_goals)):
    axes[i].title.set_text(principles_goals[i])
    columns = [key for key in data.keys() if principles_goals[i] in key ]
    sns.heatmap(data[columns].dropna(axis = 0, how = "any").corr(), cmap = "BuGn", 
                annot = True, cbar = False, xticklabels = list(range(1,len(columns)+1)), yticklabels = list(range(1,len(columns)+1)), ax = axes[i])


# Get current size

fig_size = plt.rcParams["figure.figsize"]
    
# Set figure width and height

fig_size[0] = 30
fig_size[1] = 30
plt.rcParams["figure.figsize"] = fig_size
  
plt.show()
```

```python
### Calculate Cronbach Alpha with the cleaned items ###

# The items to not consider in analysis based upon illustrated correlations

itemsToTakeOut = {"Reuse of artifacts 3", "Automation 3", "Integrated end-to-end thinking 3", 
                  "Short   cycles   and   incremental change 3",
             "Testing 3", "Monitoring 1", "Data-driven improvement 3", "Process-oriented data pipelines 3",
            "Continuous improvement 3", "Orchestration 3", "Agility and speed 3",
                  "Collaboration and trust 2"}

# Initialize values and calculate the new Cronbach Alpha

cronbachalphas2 = []
row = []
for item in principles_goals: 
    
    columns = list({key for key in data.keys() if item in key }- itemsToTakeOut)
    if len(columns) == 1:
        continue
        
    small_data = data[columns].dropna(axis = 0, how = "any")
    alpha = CronbachAlpha(small_data)
    cronbachalphas2.append(alpha)
    row.append(item)
    
pd.DataFrame.from_dict({"Items": row, "Cronbach alpha": cronbachalphas2})
```
